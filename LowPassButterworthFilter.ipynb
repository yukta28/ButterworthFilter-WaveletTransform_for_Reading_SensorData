```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

# Simulated sensor signal with noise
fs = 100  # Sampling frequency (Hz)
t = np.linspace(0, 1, fs, endpoint=False)  # Time vector
sensor_signal = np.sin(2*np.pi*5*t) + 0.5*np.random.randn(len(t))  # 5 Hz sine wave with noise

# Design a 4th-order low-pass Butterworth filter with cutoff frequency of 10 Hz
cutoff = 10  # Hz
b, a = signal.butter(N=4, Wn=cutoff/(fs/2), btype='low', analog=False)

# Apply the filter
filtered_signal = signal.filtfilt(b, a, sensor_signal)

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(t, sensor_signal, label='Noisy Sensor Data', alpha=0.5)
plt.plot(t, filtered_signal, label='Filtered Signal', linewidth=2)
plt.legend()
plt.title('Low-Pass Butterworth Filter Applied to Sensor Data')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```
